## Code for dummy data

```python
#importing data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading data
data = pd.read_csv('Book1.csv')
data.head()
```

```python
#separating training data and output lables
X_train = data.values[:,0:3]
Y_train = data.values[:,3]
```

```python
w1 = w2 = w3 = w4 = 1
L = 0.01  # The learning Rate
epochs = 1  # The number of iterations to perform gradient descent
n=4
```

```python
for i in range(epochs): 
    Y_pred = w1 + (w2*(X_train[:,0])) + (w3*(X_train[:,1]) + (w4*(X_train[:,2])))
    D_w1 = (1/n) * sum(Y_train - Y_pred)
    D_w2 = (1/n) * sum(X_train[:,0] * (Y_train - Y_pred))
    D_w3 = (1/n) * sum(X_train[:,1] * (Y_train - Y_pred))
    D_w4 = (1/n) * sum(X_train[:,2] * (Y_train - Y_pred))
    
    w1 = w1 - L * D_w1 
    w2 = w2 - L * D_w2 
    w3 = w3 - L * D_w3 
    w4 = w4 - L * D_w4                 
                       
print(w1)
print(w2)
print(w3)
print(w4)
```

```python
#importing data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading data
data = pd.read_csv('GPA.csv')
data.head()
```

```python
#separating training data and output lables
X_train = data.values[:,0]
Y_train = data.values[:,1]
```

```python
#plotting data
plt.xlabel("Hours of Study Daily")
plt.ylabel("GPA")
plt.scatter(X_train, Y_train)
plt.show()
```

```python
w1 = w2 =  1
L = 0.01  # The learning Rate
epochs = 500  # The number of iterations to perform gradient descent
n=X_train.size
```

```python
for i in range(epochs):
    #calculating Y_predicted
    Y_pred = w1 + (w2*(X_train)) 
    #finding derivtives
    D_w1 = (-2/n) * sum(Y_train - Y_pred)
    D_w2 = (-2/n) * sum(X_train * (Y_train - Y_pred))
    #updating weights
    w1 = w1 - L * D_w1 
    w2 = w2 - L * D_w2 
    
    
print("Bias:", w1)
print("Weight of x1:", w2)    
Y = sum(Y_train)
Y_p = sum(Y_pred)
print("Cost Function Value:", 1/n*((Y - Y_p)**2))
```
